```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

#Read csv with the movie ratings into a pandas df:
movie_rating_data_df = pd.read_csv("/Users/aliahsansumon/Desktop/Movies Rating.csv")

#check the df
movie_rating_data_df.head()
```

```python
#Average ratings for each user
ratings_sorted_df = movie_rating_data_df.copy()
ratings_sorted_df['User Average'] = ratings_sorted_df.mean(numeric_only = True, axis = 1)
ratings_sorted_df
```

```python
#Average ratings for each movie
movie_average_ratings_df = ratings_sorted_df.copy()
movie_average_ratings_df.loc['Movie Average'] = movie_average_ratings_df.mean(skipna=True, numeric_only=True)
movie_average_ratings_df
```

Normalization scales all the numeric values from the range [0.1]
and for standardization the data will be transformed to have a mean of 0 and a standard deviation of 1.
Normalization is a advantage in a way that it puts the data into a ver small interval so that it can be viewed clearly
and comparison/constrast can be performed easily. Disadvantage of this is that the original data will be changed and
a few information will be changed.
